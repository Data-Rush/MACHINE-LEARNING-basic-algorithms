# **MATPLOTLIB tutorial**

> Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

*   Generally easy to get started for simple plots
*   Support for custom labels and texts
* Great control of every element in a figure
*High-quality output in many formats
*Very customizable in general




```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c="b")
plt.xlabel("X axis")
plt.ylabel("Y label")
plt.title("graph in 2D") 
plt.savefig("graphfig")
```

```python
y=x*x
```

```python
plt.plot(x,y,color='green', marker='o', linestyle='dashed',linewidth = 2,markersize = 14)
```

```python
plt.plot(x,y,'r--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(0,5*np.pi,0.5)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  

plt.subplot(2, 1, 1) 
plt.plot(x, y_sin,'r--') 
plt.title('Sine') 

plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine') 
```

```python
x = [2,8,10,6,2.5] 
y = [11,16,9,1,1]  
plt.bar(x, y,color = 'r',width= 0.2) 
x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x2,y2,color = 'g')
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a , color= 'skyblue') 
plt.title("histogram") 
```

# **BOXXPLOT LIB**

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
